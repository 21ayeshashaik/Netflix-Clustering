#### Chart - 1

````python
# Chart - 1 visualization code
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Countplot for Content Type
plt.figure(figsize=(8,5))
ax = sns.countplot(x='type', data=df, palette='Set2')
plt.title('Distribution of Content Type on Netflix', fontsize=14)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Adding values on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + 0.3, p.get_height() + 50))

plt.tight_layout()
plt.show()

````

##### 1. Why did you pick the specific chart?

We chose a count plot because it effectively visualizes categorical distributions like "type" (Movies or TV Shows), helping us see which content type dominates the platform

##### 2. What is/are the insight(s) found from the chart?

The plot shows that Movies are more frequent than TV Shows on Netflix by a significant margin. This indicates that Netflix has prioritized movies over series in its content library.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. Understanding that movies dominate the platform can guide:

Content acquisition strategies to balance movie and series production.

User engagement metrics, since preferences may vary by region or age.

No negative growth insight is identified here, but an imbalance may cause Netflix to lose subscribers who prefer TV shows, suggesting the need for diversification.

#### Chart - 2

````python
# Chart - 2 visualization code
# Chart 2: Number of Titles Added Each Year on Netflix

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert date_added to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year from valid dates
df['year_added'] = df['date_added'].dt.year

# Drop NaNs and group by year
titles_by_year = df['year_added'].dropna().astype(int).value_counts().sort_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=titles_by_year.index, y=titles_by_year.values, marker='o', color='green')
plt.title('Number of Titles Added Each Year on Netflix', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

To analyze how Netflix's content library has evolved over time, especially focusing on years with rapid growth or stagnation. A line chart is ideal to observe trends over a time series

##### 2. What is/are the insight(s) found from the chart?

Netflix rapidly expanded its content between 2016 and 2020.

There was a sharp decline in 2020, likely due to production halts during the COVID-19 pandemic.

Recent years may show a plateau or gradual decline post-pandemic.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. The decline in new content addition during 2020‚Äì2021 may have impacted user satisfaction or growth in those years, especially with increased user demand during lockdowns. If not compensated with quality or engagement initiatives, it may have led to increased churn or reduced subscriber acquisition.

Justification: A drop in content variety or freshness can directly affect platform stickiness, especially in a competitive OTT market

##### 1. Why did you pick the specific chart?

To analyze how Netflix has evolved its content strategy over time, especially whether it is focusing more on TV shows or movies in recent years

##### 2. What is/are the insight(s) found from the chart?

After 2016, there‚Äôs a visible rise in TV shows, suggesting a shift toward episodic content.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Prepare dataset with numerical features only
df_corr = df.copy()

# Convert 'type' to numeric (Movie=0, TV Show=1)
df_corr['type_num'] = df_corr['type'].map({'Movie': 0, 'TV Show': 1})

# Convert 'date_added' to datetime and extract year, month
df_corr['date_added'] = pd.to_datetime(df_corr['date_added'], errors='coerce')
df_corr['add_year'] = df_corr['date_added'].dt.year
df_corr['add_month'] = df_corr['date_added'].dt.month

# Correlation matrix
corr = df_corr[['type_num', 'release_year', 'add_year', 'add_month']].corr()

# Plotting
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=14)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Identify how features like release year, added year, and content type are related.

Detect potential trends or biases (e.g., if older content is more often TV Shows).

##### 2. What is/are the insight(s) found from the chart?

Useful in feature selection for machine learning models.

Shows if Netflix is focusing more on modern vs. classic content.

Helps content strategists align release timelines with content types.



#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code
# Step 1: Ensure 'date_added' is in datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Step 2: Create 'add_year' and 'add_month' from 'date_added'
df['add_year'] = df['date_added'].dt.year
df['add_month'] = df['date_added'].dt.month

# Step 3: Convert 'type' to numeric values (Movie: 0, TV Show: 1)
df['type_num'] = df['type'].map({'Movie': 0, 'TV Show': 1})

# Step 4: Prepare dataset for the pair plot
df_pairplot = df[['type_num', 'release_year', 'add_year', 'add_month']]

# Step 5: Plotting the pair plot
sns.pairplot(df_pairplot, hue='type_num', palette='coolwarm', markers=["o", "s"])
plt.suptitle('Pair Plot of Netflix Features', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()

````

##### 1. Why did you pick the specific chart?

I chose the Pair Plot because it allows for the visualization of relationships between multiple numerical variables in the dataset. Pair plots are useful for understanding pairwise correlations, distributions, and potential clustering between features.

##### 2. What is/are the insight(s) found from the chart?

Separation by Type: The pair plot clearly shows the separation between Movies and TV Shows (represented by different colors) for the features like release_year, add_year, and add_month. This indicates that Movies and TV Shows have distinct patterns in these variables.
